In thanksgiving_data.csv the columns are shifted by one position on the right. We solve it by using the `iloc` method. 

DataFrame.iloc: Purely integer-location based indexing for selection by position.

```python
thanks_bad = pd.read_csv(os.path.join("data", "thanksgiving_data.csv"))
thanks_bad.head()
```

```python
col = thanks_bad.columns # we select the columns labels
thanks = thanks_bad.iloc[:, 1:] # we remove the first column
thanks.columns = col[0:-1] # we set the columns removing the last one
thanks.head()
```

## Series and DataFrames
Tables are stored in DataFrames, and each column of a DataFrame is a Series object, that is, basically a numpy array supporting mixed data types (casted to object type). 

There are different ways to create them.

```python
s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series(['a', 'b', 'c', 'd', 'e'])
s3 = pd.Series([1, 'a', 1.0])
```

```python
s4 = pd.Series([1, 2])
s5 = pd.Series(['a', 'b'])
print(s4)
print(s5)
pd.DataFrame([s4, s5], index=["c", "d"])
```

```python
# We can create a DataFrame from a dictionary
d = {"president": ["Salvini","Di Maio","Paperino","Maga Mago"], "chance": [1,2,8,12] }
df = pd.DataFrame(d,columns=["president","chance"])
df
```
